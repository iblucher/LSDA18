```python
import numpy as np
import pickle
import csv
from sklearn.tree import DecisionTreeClassifier
```

```python
def create_subset():

    # Create random subsets from training set
    n = 50000000
    M = 15000

    tid = sorted(np.random.choice(n, M, replace = True))
    
    u, count = np.unique(tid, return_counts = True)
    print(count.shape)
    
    subset = []
    
    cnt = 0
    count_index = 0
   
    with open('landsat_train.csv') as fd:
        reader  = csv.reader(fd)
        for idx, row in enumerate(reader, 0):
            if idx == tid[cnt]:
                if(count[count_index] > 1):
                    for i in range(count[count_index]):
                        subset.append(row)
                        cnt += 1
                else:
                    subset.append(row)
                    cnt += 1
                count_index += 1
            if cnt == 15000:
                break
            if idx % 1000000 == 0:
                print(idx)
        subset = np.array(subset)
    return subset
    
```

```python
def create_classifier(i):
    clf = DecisionTreeClassifier(criterion = 'gini', max_depth = None, min_samples_split = 2, max_features = 5, random_state = i)
    return clf
```

```python
classifiers = []

for i in range(10):
    print(i + 1)
    subset = create_subset()
    print(subset.shape)
    clf = create_classifier(i + 1)
    clf.fit(subset[:, 1:9], subset[:, 0])
    classifiers.append(clf)
```

```python
pickle.dump(classifiers, open('model.save', 'wb'))
```
